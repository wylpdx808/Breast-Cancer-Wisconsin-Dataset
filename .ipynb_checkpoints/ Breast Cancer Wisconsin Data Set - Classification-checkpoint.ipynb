**Variable descriptions: **
* 1) ID number 
* 2) Diagnosis (M = malignant, B = benign) 3-32)

There are 3 measurements (**mean**, **se**, **worst**) for each of the 10 real-valued features  computed for each cell nucleus:

* a) radius (mean of distances from center to points on the perimeter) 
* b) texture (standard deviation of gray-scale values) 
* c) perimeter 
* d) area 
* e) smoothness (local variation in radius lengths) 
* f) compactness (perimeter^2 / area - 1.0) 
* g) concavity (severity of concave portions of the contour) 
* h) concave points (number of concave portions of the contour) 
* i) symmetry 
* j) fractal dimension ("coastline approximation" - 1)

The **mean**, **standard error** and "**worst**" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features.

```python
# Separate 'diagnosis' output and drop 'id', and 'Unnamed: 32' from input
y = data.diagnosis
x = data.drop(['id','diagnosis','Unnamed: 32'], axis = 1)
```

### Scale Data 

```python
# Standardize by removing mean and scaling by unit variance 
stdsc = StandardScaler()
noscaled_x = x
x = pd.DataFrame(stdsc.fit_transform(x), columns = x.columns)
x.head()
```

#### Mean features 

```python
data_m = pd.concat([y, x.iloc[:, :10]], axis = 1)
data_m.head()
```

```python
data_m_melt = pd.melt(data_m, 
                 id_vars = 'diagnosis', 
                 var_name = 'features', 
                 value_name = 'value')
data_m_melt.head()
```

#### SE features 

```python
data_se = pd.concat([y, x.iloc[:, 10:20]], axis = 1)
data_se.head()
```

```python
data_se_melt = pd.melt(data_se, 
                  id_vars = 'diagnosis', 
                  var_name = 'features', 
                  value_name = 'value')
data_se_melt.head()
```

```python
data_w = pd.concat([y, x.iloc[:, 20:]], axis = 1)
data_w.head()
```

```python
data_w_melt = pd.melt(data_w, 
                 id_vars = 'diagnosis', 
                 var_name = 'features', 
                 value_name = 'value')
data_w_melt.head()
```

```python
data_all = pd.concat([data_m_melt, data_se_melt, data_w_melt], axis = 1)
data_all.head()
```

```python
fig, axList = plt.subplots(nrows = 3, ncols = 1)
axList = axList.flatten()
fig.set_size_inches (10,25)

for ax in enumerate(axList): 
    loc = ax[0]
    ax = ax[1]
    i_beg = loc*3
    i_end = loc*3 + 3
    
    sns.violinplot(x = 'features', 
                   y = 'value', 
                   hue = 'diagnosis', 
                   data = data_all.iloc[:, i_beg:i_end], 
                   ax = ax, 
                   inner = 'quart',
                   split = True)
    
    plt.sca(ax)
    plt.xticks(rotation = 45)
    
    titles = ['Mean', 'SE', 'Worst']
    ax.set_title(str(titles[loc]), fontsize = 20)
    
    plt.tight_layout()
```

```python
sns.set_context('talk', rc = {'axes.labelsize': 20, 'legend.numpoints': 2})
sns.pairplot(data_m, hue = 'diagnosis', size = 3);
```
